이게 되는 이유, 정규분포는 애초에 히스토그램이다. 구간은 200, 200개의 구간 (ex 0~50 까지 1bin, 1bin에 y 할당), alpha 는 투명도 인수.

```python
from scipy.stats import norm

data = norm.rvs(10.0, 3, size = 1000)
plt.hist(data, bins=20, density=True, alpha=0.6, color='b')

mu, std = norm.fit(data)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'r', linewidth=2)
```

논리가 조금 어렵다. 히스토그램 생성 -> 히스토그램의 평균, 표준편차 학습(fit) -> 히스토그램 x의 최대,최소 값 사이에 등간격으로 100개 x 생성. 여기 x 각 값에 대응하여 학습한 평균, 표준편차의 정규분포에서 대응하는 값(p : 확률밀도)을 가지는 정규 분포 생성. -> x, p 설정해 정규분포 생성.

값이 100개인데 왜 선이 연속적일까? 주어진 x 값들에 대응 하는 p 값들을 선으로 이은 선인 것이다. 이는 x가 연속적인 경우에 가능, 어떻게 보면 회귀식 같기도하다. x1 = 1 의 p1= 1 x2 = 2 의 p2 = 2 라고 가정. x1과 x2 사이의 값은 없지만 그래프에 의해 그 값에 대한 p값이 추정이 가능하다.
